# Exploratory Data Analysis

### Gender

```python
gen = sns.countplot(x = 'Gender', data = df)
for bars in gen.containers:
    gen.bar_label(bars)
```

```python
gen_amount = df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc = {'figure.figsize': (5,3)})
sns.barplot(x = 'Gender', y = 'Amount', data =gen_amount)
```

#### From the above graph we can see that most buyers are female and their purchasing power is higher than male.

### Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
sns.set(rc = {'figure.figsize': (10,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount vs age group
amount_age = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x = "Age Group", y = 'Amount', data = amount_age)
```

#### From the above graph we can see that most buyers are from age group 26-35.

### State

```python
state_order = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(x = 'State', y = 'Orders', data = state_order)
```

```python
state_amount = df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
sns.set(rc = {'figure.figsize':(15,5)})
sns.barplot(x = 'State', y = 'Amount', data = state_amount)
```
